```python
P0=120
fig, ax=plt.subplots()
ax.grid()
Q0=500
wW=lL
for deltd, PP0 in dic.items():
    wwW=(lL*1/((1-deltd)**4))
    Pab=(P0/(2+(wW)))
    #Pab=Q0*W
    #Pbc=np.array((P0*ww)/(2*W+w))
    Pbc=(P0*(wwW))/(2+wW)
    #Pbc=Q0*ww
    print(PP0)
    Pcd=Pab
    P1=PP0-Pab
    P2=P1-Pbc
    P3=P2-Pcd
    li=[Pab, Pbc, Pcd]
    plt.plot(['a','b','c','d'], [PP0, P1, P2, P3])
    print(li)
plt.ylabel('P')
plt.legend(labels=['d/D={}'.format(dD[0]), 'd/D={:.3f}'.format(dD[1]), 'd/D={:.3f}'.format(dD[2])])
```

## Завдання №2
### Частина 1

```python
fig, ax=plt.subplots()
ax.grid()
Pa=80
Pd=50
lL=0.2+theta
dD=np.array([0.1+theta, 0.3+theta, 0.7+theta])
wW=lL
for d in dD:
    wwW=lL*(1/(1-d)**4)
    Pab=(Pa-Pd)/(2+wwW)
    Pcd=Pab
    Pbc=((Pa-Pd)*wwW)/(2+wwW)
    a=Pa
    b=Pa-Pab
    c=b-Pbc
    d=c-Pcd
    plt.plot(['a','b','c','d'], [a, b, c, d])
    print(Pab, Pbc)
plt.ylabel('P')
plt.legend(labels=['d/D={:.3f}'.format(dD[0]), 'd/D={:.3f}'.format(dD[1]), 'd/D={:.3f}'.format(dD[2])])
```

## Висновок
Тиск в точках a та b практично не змінюється через те, що потік крові стикаючись з звуженням сповільнується і тиск падає повільніше. Фільтраційно-реабсорційні процеси покращаться перед тромбом через збільшення тиску, і погіршаться безпосередньо після тромбу. Розташування тромбу на цей ефект не впливає.

### Частина 2

```python
fig, ax=plt.subplots()
ax.grid()
lL=np.array([0.05, 0.1+theta, 0.2+theta, 0.4+theta, 0.8])
for l in lL:
    dD=np.linspace(0,0.9999,100)
    wW=l
    wwW=l*(1/(1-dD)**4)
    n=10
    Q0=100
    q0=Q0/n
    qq=q0*((2+wW)/(2+wwW))
    qq0=qq/q0
    plt.plot(dD, qq0)
plt.ylabel('q`/q$_0$')
plt.xlabel('d/D')
plt.legend(labels=['l/L={:.3f}'.format(lL[0]), 'l/L={:.3f}'.format(lL[1]), 'l/L={:.3f}'.format(lL[2]),
                  'l/L={:.3f}'.format(lL[3]), 'l/L={:.3f}'.format(lL[4])])
```

## Висновок
З побудованого графіку бачимо, що збільшені ширини області звуження, швидше спадає швикість кровоку зі збільшенням тромбу. І навпаки при малій області звуження, швиткість кровотоку достатньо дового тримається близько початкового рівня, хоча при подальшому збільшенні тромбу спостерігається різкий спад швидкості.

```python
N=[0.015, 0.005, 0.0025]
Q=1
R=1
for n in N:
    gradP=n*((8*Q)/(np.pi*R**4))
    plt.plot([0,1],[80, 80-gradP])
plt.tick_params(left = False, right = False , labelleft = False ,
                labelbottom = False, bottom = False)
plt.ylabel('P')
plt.xlabel('X')
plt.legend(labels=['\u03B7={}'.format(N[0]), '\u03B7={}'.format(N[1]), '\u03B7={}'.format(N[2])])
```
