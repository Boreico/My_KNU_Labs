```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
theta=0.1/4
```

## Завдання №1

Створюємо і тестуємо модель способу введення ліків - "ін'єкція"

```python
t=np.linspace(0, 10, 100)
koef=[0.1, 0.2, 0.3]
fig, ax=plt.subplots(figsize=(4, 3))
ax.grid()
for k in koef:
    def model_1(m, t):
        dmdt=-k*m
        return dmdt
    m=odeint(model_1, 5, t)
    plt.plot(t, m)
```

Вдалося відтворити Рис.4 з Лабораторної роботи, тобто модель працює.

Будуємо графіки систем з завдання:

```python
m0=3+theta
t=np.linspace(0, 20, 100)
koef=sorted([0.20, 0.10, 0.05])
fig, ax=plt.subplots(figsize=(4, 3))
ax.grid()
for k in koef:
    def model_1(m, t):
        dmdt=-k*m
        return dmdt
    m=odeint(model_1, m0, t)
    plt.plot(t, m)
plt.legend(['k={:.2f}'.format(koef[0]), 'k={:.2f}'.format(koef[1]), 'k={:.2f}'.format(koef[2])])
```

$T_{0.5}$ для $k=0.05$ рівне 14, для $k=0.1$ рівне 7, а для $k=0.2$ рівне 3.5.

Розраховуємо параметри $m_0$ та $Q$, при $k=0.2+\Theta$, щоб відразу встановлювалась оптимальна маса ліків в крові.

```python
k=0.2+theta
m_opt=np.array([2+theta, 4+theta, 6+theta])
m0=m_opt
t=np.linspace(1, 20, 100)
Q=m_opt*k
fig, ax=plt.subplots(figsize=(4, 3))
ax.grid()
d={Q[0]:m0[0], Q[1]:m0[1], Q[2]:m0[2]}
for q, m00 in d.items():
    def model_2(m, t):
        dmdt=q-k*m
        return dmdt
    m=odeint(model_1, m00, t)
    plt.plot(t, m)

plt.legend(['Q={:.2f}, $m_0$={:.2f}'.format(Q[0], m0[0]), 'Q={:.2f}, $m_0$={:.2f}'.format(Q[1], m0[1]), 'Q={:.2f}, $m_0$={:.2f}'.format(Q[2], m0[2])])
```
