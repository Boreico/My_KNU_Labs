```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
theta=0.1/4
theta
```

```python
def Px(P, x, P0, lambd):
    return [P[1], (P[0]/(lambd**2))-P0/(lambd**2)]
```

```python
def Px_full(P0, Pb, Pa, L, lambd, x):
    return (((P0-Pb)+(Pa-P0)*np.e**(-L/lambd))/(np.e**(-L/lambd)-np.e**(L/lambd)))*(np.e**(x/lambd))+(((Pb-P0)+(P0-Pa)*np.e**(L/lambd))/(np.e**(-L/lambd)-np.e**(L/lambd)))*(np.e**(-x/lambd))+P0
```

```python
def q(P, P0, w):
    return (P-P0)/w
```

```python
Pa=np.array([28.4+theta, 31+theta, 33+theta])*133.322
Pb=12*133.322
P0=20*133.322
L=600/1000000
l=0.6/1000000
R=3/1000000
N=1.3*(10**12)
n=0.0012
r=10/1000000000
W=(8*n)/(np.pi*(R**4))
w=(4*n*l)/((np.pi**2)*(r**4)*R*N)
lambd=np.sqrt(w/W)
#lambd=np.sqrt(((R**3)*l)/(2*np.pi*N*(r**4)))
for pa in Pa:
    y=q(Px_full(P0, Pb, pa, L, lambd, x), P0, w)
    plt.plot(x, y)
plt.legend(labels=['$P_a={:.1f}$ Pa'.format(Pa[0]), '$P_a={:.1f}$ Pa'.format(Pa[1]), '$P_a={:.1f}$ Pa'.format(Pa[2])])
plt.ylabel('q')
plt.xlabel('x')
```

```python
def Q(P, W):
    return (-1/W)*np.gradient(P)
```

```python
Pa=np.array([28.4+theta, 31+theta, 33+theta])*133.322
Pb=12*133.322
P0=20*133.322
L=600/1000000
l=0.6/1000000
R=3/1000000
N=1.3*(10**12)
n=0.0012
r=10/1000000000
W=(8*n)/(np.pi*(R**4))
w=(4*n*l)/((np.pi**2)*(r**4)*R*N)
#lambd=np.sqrt(w/W)
lambd=np.sqrt(((R**3)*l)/(2*np.pi*N*(r**4)))
for pa in Pa:
    y=Q(Px_full(P0, Pb, pa, L, lambd, x), W)
    plt.plot(x, y)
plt.legend(labels=['$P_a={:.1f}$ Pa'.format(Pa[0]), '$P_a={:.1f}$ Pa'.format(Pa[1]), '$P_a={:.1f}$ Pa'.format(Pa[2])])
plt.ylabel('Q')
plt.xlabel('x')
```

```python
def K(q, L, Xa):
    return (quad(q, 0, L))/(quad(q, 0, Xa))
```

```python
Pa=(28.4+theta)*133.322
Pb=12*133.322
P0=20*133.322
L=600/1000000
l=0.6/1000000
R=3/1000000
N=1.3*(10**12)
n=0.0012
r=10/1000000000
W=(8*n)/(np.pi*(R**4))
w=(4*n*l)/((np.pi**2)*(r**4)*R*N)
lambd=np.sqrt(w/W)
#lambd=np.sqrt(((R**3)*l)/(2*np.pi*N*(r**4)))
for xx in x:
    if q(Px_full(P0, Pb, Pa, L, lambd, xx), P0, w) < 0:
        Xa=xx
        break
pxf=Px_full(P0, Pb, Pa, L, lambd, x)
K=K(q, L, Xa, (pxf, P0, w))
print('K={}'.format(K))
```
